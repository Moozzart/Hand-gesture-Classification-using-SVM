```python
#Function to convert RGB image to Grayscale
def rgb_to_grey(s,i):
    img = Image.open('C:\\Users\\Utkarsh Pratap Singh\\Lab6\\Lab0609'+'\\'+s+'\\'+s+str(i)+'.jpg').convert('L')
    img.save('grey'+s+str(i)+'.jpg')
```

```python
#To reduce the size of image from 200x200 to 25x25 for mitigating the Memory Error
def reduce_size(s,i):
    img = Image.open('C:\\Users\\Utkarsh Pratap Singh\\Lab6\\Lab0609'+'\\grey'+s+str(i)+'.jpg')
    img=img.resize((25,25))
    img.save('grey'+s+str(i)+'.jpg')
```

```python
#Function that takes the image pixels matrix without flattening 
#threshold is given in the 'If' conditional statement below and returns the matrix
def thresholding1(imgarr): #input 200x200 array
    for i in range(len(imgarr)):
        for j in range(len(imgarr[0])):
            if (imgarr[i][j]<200 and imgarr[i][j]>30):
                imgarr[i][j]=255
            else:
                imgarr[i][j]=0
    return imgarr
```

```python
#Function that takes the image pixels matrix with flattening(turning it into an array)
#threshold is given in the 'If' conditional statement below and return the array
def thresholding(imgarr): #input flatten vector
    for i in range(len(imgarr)):
        if ((imgarr[i]<200) and (imgarr[i]>30)):
            imgarr[i]=255
        else:
            imgarr[i]=0
    return imgarr
```

```python
#Function for Subtracting the Background from the Image
def grey_without_back(s,i):   
    im = Image.open('C:\\Users\\Utkarsh Pratap Singh\\Lab6\\Lab0609'+'\\grey'+s+str(i)+'.jpg')
    ba = Image.open('C:\\Users\\Utkarsh Pratap Singh\\Lab6\\Lab0609'+'\\'+'greynothing'+str(i)+'.jpg')
    I=np.asarray(im)
    B=np.asarray(ba)
    d=I-B
    d=d.flatten()
    return thresholding(d)
```

## Now implementing GridSearchCV to get the best possible combination of parameters among the given parameters

```python
#Fitting GridSearchCV on trainig set 
gridsearch=GridSearchCV(classifier,params,refit = True, verbose = 3)
gridsearch.fit(X_train, y_train)
```
